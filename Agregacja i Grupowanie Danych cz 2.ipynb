## Wczytanie i podgląd danych

```python
import pandas as pd

superstore_df = pd.read_excel('Superstore.xlsx', sheet_name='Orders')
superstore_df.head()
```

## Przykłady zastosowania grupowania i agregacji

### Łączna sprzedaż w każdym regionie

```python
region_sales = superstore_df.groupby('Region')['Sales'].sum()
region_sales
```

### Liczba zamówień i pozycji w każdej podkategorii

```python
subcategory_orders = superstore_df.groupby('Sub-Category')['Order ID'].agg(['count', 'nunique'])
subcategory_orders.columns = ['Total Orders', 'Unique Orders']
subcategory_orders
```

### Średni zysk na produkt w każdej kategorii

```python
avg_profit_per_product = superstore_df.groupby(['Category', 'Product Name'])['Profit'].mean().reset_index()
avg_profit_per_product
```

### Top 5 produktów z największą sprzedażą w każdej kategorii

```python
grouped_df = superstore_df.groupby(['Category', 'Product Name'])['Sales'].sum().reset_index()

grouped_df
```

```python
top_5_products = grouped_df.sort_values(['Category', 'Sales'], ascending=[True, False]).groupby('Category').head(5)
top_5_products
```

```python
# II sposób
top_5_products_II = grouped_df.groupby('Category').apply(lambda x: x.nlargest(5, 'Sales')).reset_index(drop=True)
top_5_products_II
```

## Seaborn

Seaborn jest zbudowany na bazie Matplotlib, ale oferuje bardziej zintegrowane funkcje do pracy z danymi w Pandas DataFrame oraz dodatkowe style i palety kolorów.

```python
# Na początek zaimportujemy Seaborn oraz inne potrzebne biblioteki.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Dane do naszych przykładów
df = pd.DataFrame({
    'OrderDate': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05', 
                  '2024-01-06', '2024-01-07', '2024-01-08', '2024-01-09', '2024-01-10'],
    'Category': ['Electronics', 'Furniture', 'Furniture', 'Electronics', 'Clothing', 
                 'Clothing', 'Electronics', 'Furniture', 'Clothing', 'Furniture'],
    'Sales': [200, 450, 300, 500, 150, 400, 700, 300, 200, 350],
    'Quantity': [2, 5, 3, 4, 1, 4, 6, 3, 2, 5]
})
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df.head()
```

## Wykres słupkowy (Bar Plot) w Seaborn

Seaborn umożliwia tworzenie wykresów słupkowych z wbudowanym grupowaniem i agregacją. Możesz szybko stworzyć wykres, podając kolumny z DataFrame.

Porównując, w Matplotlib musiałbyś najpierw wykonać agregację, a następnie stworzyć wykres.

```python
sns.barplot(x='Category', y='Sales', data=df, errorbar=None)
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

## Wykres punktowy (Scatter Plot) w Seaborn

Seaborn automatycznie obsługuje różne aspekty estetyczne, takie jak kolorowanie punktów według kategorii, oraz dodaje linie regresji, co jest szczególnie użyteczne w analizie danych.

W Matplotlib potrzebowałbyś dodatkowych kroków, aby osiągnąć ten sam poziom estetyki, np. ręczne przypisanie kolorów do kategorii.

```python
sns.scatterplot(x='Sales', y='Quantity', hue='Category', data=df)
plt.title('Zależność między Sprzedażą a Ilością')
plt.xlabel('Sprzedaż')
plt.ylabel('Ilość')
plt.show()
```

## Wykres rozkładu (Distribution Plot) w Seaborn

Seaborn oferuje łatwe narzędzia do tworzenia wykresów rozkładu z gęstością jądra (KDE). W Matplotlib musiałbyś osobno tworzyć histogram oraz wykres gęstości.

KDE działa tak, że "rozsiewa" wartości danych na małe "górki" (krzywe Gaussa), a następnie sumuje te górki, aby uzyskać gładką krzywą. W rezultacie, zamiast widzieć sztywne przedziały, widzisz, gdzie najwięcej danych jest skoncentrowanych.

```python
sns.histplot(df['Sales'], kde=True)
plt.title('Rozkład Sprzedaży')
plt.xlabel('Sprzedaż')
plt.ylabel('Częstotliwość')
plt.show()
```

## Dostosowywanie wyglądu wykresów

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    'Dates - Days': [1, 2, 3, 4, 5],
    'Sales Values': [10, 15, 13, 17, 19]
})
```

```python
x, y = 'Days', 'Sales'

df = df.rename(columns={'Dates - Days': x, 'Sales Values': y})
df
```

```python
plt.plot(df[x], df[y], color='red', linestyle = '--')
plt.xlabel(x)
plt.xticks(ticks = df[x], labels=df[x].astype(int))
plt.ylabel(y)
plt.yticks(ticks = range(0, df[y].max() + 10, 5))
plt.grid(True)
plt.show()
```

## Tworzenie wykresów z wieloma liniami

Możemy na jednym wykresie umieścić kilka serii danych, korzystając z metody plot() wielokrotnie.

```python
# Przykładowe dane
df = pd.DataFrame({
    'Day': [1, 2, 3, 4, 5],
    'Sales1': [10, 15, 13, 17, 19],
    'Sales2': [7, 21, 20, 14, 16]
})
```

```python
plt.plot(df['Day'], df['Sales1'], color='red', linestyle = '--', label='Sales1')
plt.plot(df['Day'], df['Sales2'], color='blue', linestyle = '-', label='Sales2')
plt.xlabel('Day')
plt.xticks(ticks = df['Day'], labels=df['Day'].astype(int))
plt.ylabel('Sales')
plt.yticks(ticks = range(0, df['Sales2'].max() + 10, 5))
plt.legend()
plt.grid(True)
plt.show()
```

## Zapisywanie wykresów

Matplotlib umożliwia zapisywanie wykresów do plików w różnych formatach, takich jak PNG, PDF, SVG itp.

```python
plt.plot(df['Day'], df['Sales1'])
plt.title('Wykres liniowy do zapisu do pliku')
plt.xlabel('Dzień')
plt.ylabel('Sprzedaż')
plt.grid(True)
plt.savefig('line_plot.png')
plt.savefig('line_plot.pdf')
plt.show()
```

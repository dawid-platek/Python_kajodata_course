Poniżej znajdziesz 3 ćwiczenia, które pozwolą Ci przećwiczyć podstawy Matplotlib. Każde ćwiczenie zawiera zadanie oraz rozwiązanie, które możesz wypróbować samodzielnie.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Ćwiczenie 1: Wykres Słupkowy

Stwórz wykres słupkowy przedstawiający liczbę sprzedanych produktów w różnych kategoriach, używając poniższych danych

```python
df1 = pd.DataFrame({
    'Category': ['Electronics', 'Furniture', 'Clothing', 'Food'],
    'Sales': [150, 200, 120, 170]
})
```

```python
plt.bar(df1['Category'], df1['Sales'], color = 'green')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

## Ćwiczenie 2: Wykres Liniowy

Używając poniższych danych, stwórz wykres liniowy przedstawiający zmianę temperatury w ciągu tygodnia. Dodaj tytuł wykresu oraz etykiety osi X i Y. Dodaj marker 'o', ustaw styl linii jako '-', a kolor jako niebieski. Do wykresu dodaj siatkę. Dzień tygodnia na osi opisz przez pierwsze dwie litery tego dnia (**podpowiedź - użyj "str[:2]"**).

```python
df2 = pd.DataFrame({
    'Day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
    'Temperature': [22, 21, 23, 24, 22, 25, 24]
})
```

```python
plt.plot(df2['Day'], df2['Temperature'], marker='o', linestyle='-', color='blue')
plt.title('Temperature in the week')
plt.xlabel('Day')
plt.xticks(ticks=df2['Day'], labels = df2['Day'].str[:2])
plt.ylabel('Temperature')
plt.grid(True)
plt.show()
```

## Ćwiczenie 3: GroupBy

Używając poniższych danych, wykonaj agregację, która pokaże średnią ilość sprzedanych produktów (Quantity) w każdej kategorii (Category). Następnie stwórz wykres słupkowy przedstawiający te wartości.Dodaj siatkę do wykresu.

```python
df3 = pd.DataFrame({
    'Category': ['Electronics', 'Furniture', 'Clothing', 'Food', 'Electronics', 'Furniture', 'Clothing', 'Food'],
    'Quantity': [10, 5, 8, 12, 15, 7, 6, 10]
})
```

```python
df3.groupby('Category')['Quantity'].mean().plot(kind='bar')
plt.ylabel('Average Quantity')
plt.xlabel('Category')
plt.grid(True)
plt.show()
```

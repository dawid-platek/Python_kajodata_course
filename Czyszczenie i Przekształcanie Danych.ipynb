W tej lekcji skupimy się na czyszczeniu i przekształcaniu danych w Pandas, korzystając z pliku Excelowego "shark attacks.xlsx". 

Zajmiemy się przekształcaniem i czyszczeniem tekstu, radzeniem sobie z wartościami null oraz duplikatami.

```python
import pandas as pd 

shark_df = pd.read_excel('shark attacks.xlsx')
shark_df.head(10)
```

```python
shark_df.tail(10)
```

```python
shark_df.info()
```

## Przekształcanie i czyszczenie tekstu

### Usuwanie zbędnych spacji z kolumn tekstowych

```python
shark_df['Species'] = shark_df['Species'].str.strip()

for col in shark_df.select_dtypes(include='object').columns:
    shark_df[col] = shark_df[col].str.strip()
```

### Zamiana tekstu na małe litery

```python
cols = [col for col in shark_df.select_dtypes(include=['object']).columns if col != 'Investigator or Source']
print(cols)

for c in cols:
    shark_df[c] = shark_df[c].str.lower()

shark_df.head()
```

### Zamiana określonych wartości na NaN

```python
import numpy as np 
shark_df.info()

for col in shark_df.select_dtypes(include='object').columns:
    shark_df[col] = shark_df[col].replace(['unknown', ''], np.nan)
shark_df.info()
```

### Sprawdzenie unikalnych wartości

```python
for c in ['Date', 'Sex', 'Fatal (Y/N)']:
    print(c)
    print(shark_df[c].unique().tolist())
    print('-----------------')
```

### Zamiana określonych wartości cz. 2 :D

#### Praca z datami

```python
shark_df['Date'] = shark_df['Date'].str.replace('reported ', '').str.replace('--', '-').str.strip()
shark_df['Date_2'] = pd.to_datetime(shark_df['Date'], errors='coerce')
shark_df.info()
```

```python
shark_df[['Date', 'Date_2']][shark_df['Date'].notna()].head(10)
```

#### Zamiana wartości na podstawie warunku

```python
shark_df['Fatal (1/0)'] = shark_df['Fatal (Y/N)'].apply(lambda x: 1 if x == 'y' else 0)

print(shark_df[['Fatal (Y/N)', 'Fatal (1/0)']][shark_df['Fatal (Y/N)'] == 'n'].head(3))
print(shark_df[['Fatal (Y/N)', 'Fatal (1/0)']][shark_df['Fatal (Y/N)'] == 'y'].head(3))
```

```python
shark_df['Sex'] = shark_df['Sex'].apply(lambda x: x if x in ['m', 'f'] else np.nan)

print(shark_df['Sex'].unique().tolist())
```

### Radzenie sobie z duplikatami

```python
print(shark_df.duplicated().sum())
```

```python
shark_df.tail(5)
```

```python
shark_df.drop_duplicates(inplace=True)
shark_df.info()
shark_df.tail(5)
```

```python
shark_df.reset_index(drop=True, inplace=True)
shark_df.tail(5)
```

### Usuwanie NULLi (NaN / NaT)

```python
shark_df = shark_df.dropna(subset=['Case Number'])
shark_df.tail(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    'OrderDate': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05', 
                  '2024-01-06', '2024-01-07', '2024-01-08', '2024-01-09', '2024-01-10'],
    'Category': ['Electronics', 'Furniture', 'Furniture', 'Electronics', 'Clothing', 
                 'Clothing', 'Electronics', 'Furniture', 'Clothing', 'Furniture'],
    'Sales': [200, 450, 300, 500, 150, 400, 700, 300, 200, 350],
    'Quantity': [2, 5, 3, 4, 1, 4, 6, 3, 2, 5]
})
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
```

## Wykres Violinowy (Violin Plot)

Wykres Violinowy to kombinacja wykresu pudełkowego i wykresu KDE. Pokazuje rozkład danych oraz ich gęstość w różnych grupach, co jest bardzo pomocne w analizie danych, szczególnie gdy chcesz zobaczyć zarówno medianę, jak i rozkład danych.

```python
x, y = 'Category', 'Sales'
sns.violinplot(x=x, y=y, data=df, inner='quartile', palette='Set2')
plt.xlabel(x)
plt.ylabel(y)
plt.show()
```

```python
df.sort_values(by=x)
```

```python
sns.violinplot(x=x, y=y, data=df, inner='quartile', palette='Set2')
plt.xlabel(x)
plt.ylabel(y)
plt.ylim(0, df[y].max())
plt.yticks([250, 500, 750], ['Low', 'Mid', 'High'])
plt.grid(True)
plt.show()
```

## Wykres pudełkowy (Box Plot) w Seaborn

Seaborn umożliwia łatwe tworzenie wykresów pudełkowych, które pomagają zidentyfikować wartości odstające oraz rozkład danych w grupach. W Matplotlib stworzenie wykresu pudełkowego wymaga nieco więcej kodu, a formatowanie go może być bardziej pracochłonne.

```python
x, y = 'Category', 'Sales'
sns.boxplot(x=x, y=y, data=df, palette='Set2')
plt.title(x + ' vs ' +  y + ' Boxplot')
plt.xlabel(x)
plt.ylabel(y)
plt.ylim(0, df[y].max() + 250)
plt.yticks([0, 250, 500, 750], ['None', 'Low', 'Mid', 'High'])
plt.grid(True)
plt.show()
```

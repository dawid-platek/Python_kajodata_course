W tej lekcji nauczymy się, jak importować i eksportować dane w Pandas. 

Skorzystamy z 2 plików, które znajdziesz w załączniku: orders.csv oraz Superstore.xlsx

### Importowanie danych

```python
import pandas as pd

orders_df = pd.read_csv('orders.csv', sep=';')

orders_df.head()
```

```python
superstore_df = pd.read_excel('Superstore.xlsx', sheet_name = 'Orders')
superstore_df.head()
# print(superstore_df.head())
```

### Eksportowanie danych

```python
orders_df.to_csv('exported_orders.csv', index=False, sep=';')
```

```python
superstore_df.to_excel('exported_superstore.xlsx', sheet_name='Orders', index=False)
```
